```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
class Joint:
    def __init__(self, a, d, alpha):
        self.a = a
        self.d = d
        self.alpha = alpha


joint1 = Joint(0, 0.333, 0)
joint2 = Joint(0, 0, -np.pi/2)
joint3 = Joint(0, 0.316, np.pi/2)
joint4 = Joint(0.0825, 0, np.pi/2)
joint5 = Joint(-0.0825, 0.384, -np.pi/2)
joint6 = Joint(0, 0, np.pi/2)
joint7 = Joint(0.088, 0, np.pi/2)
Flange = Joint(0, 0.107, 0)
```

```python
def DH(theta, alpha, d, a):
    st = np.sin(theta)
    ct = np.cos(theta)
    sa = np.sin(alpha)
    ca = np.cos(alpha)

    matrika = np.array([[ct, -st*ca, st*sa, a*ct],
                          [st, ct*ca, -ct*sa, a*st],
                          [0, sa, ca, d],
                          [0, 0, 0, 1]])

    return matrika
```

```python
def plotting(theta1, theta2, theta3, theta4, theta5, theta6, theta7):
    
    plt.rcParams["figure.figsize"] = [15, 15]
    plt.rcParams["figure.autolayout"] = True
    
    T01 = DH(theta1, joint1.alpha, joint1.d, joint1.a)
    T12 = DH(theta2, joint2.alpha, joint2.d, joint2.a)
    T23 = DH(theta3, joint3.alpha, joint3.d, joint3.a)
    T34 = DH(theta4, joint4.alpha, joint4.d, joint4.a)
    T45 = DH(theta5, joint5.alpha, joint5.d, joint5.a)
    T56 = DH(theta6, joint6.alpha, joint6.d, joint6.a)
    T67 = DH(theta7, joint7.alpha, joint7.d, joint7.a)
    
    t1 = T01
    t2 = np.matmul(T01,T12)
    t3 = np.matmul(T01,T12,T23)
    t4 = T01@T12@T23@T34
    t5 = T01@T12@T23@T34@T45
    t6 = T01@T12@T23@T34@T45@T56
    t7 = T01@T12@T23@T34@T45@T56@T67
    
    x1,y1,z1 = t1[:-1,3:]
    x2,y2,z2 = t2[:-1,3:]
    x3,y3,z3 = t3[:-1,3:]
    x4,y4,z4 = t4[:-1,3:]
    x5,y5,z5 = t5[:-1,3:]
    x6,y6,z6 = t6[:-1,3:]
    x7,y7,z7 = t7[:-1,3:]
    
    x1 = x1[0]
    x2 = x2[0]
    x3 = x3[0]
    x4 = x4[0]
    x5 = x5[0]
    x6 = x6[0]
    x7 = x7[0]
    
    y1 = y1[0]
    y2 = y2[0]
    y3 = y3[0]
    y4 = y4[0]
    y5 = y5[0]
    y6 = y6[0]
    
    z1 = z1[0]
    z2 = z2[0]
    z3 = z3[0]
    z4 = z4[0]
    z5 = z5[0]
    z6 = z6[0]
    z7 = z7[0]    
    
    crta1x = [x1,x2]
    crta2x = [x2,x3]
    crta3x = [x3,x4]
    crta4x = [x4,x5]
    crta5x = [x5,x6]
    crta6x = [x6,x7]
    
    crta1y = [y1,y2]
    crta2y = [y2,y3]
    crta3y = [y3,y4]
    crta4y = [y4,y5]
    crta5y = [y5,y6]
    crta6y = [y6,y7]
    
    crta1z = [z1,z2]
    crta2z = [z2,z3]
    crta3z = [z3,z4]
    crta4z = [z4,z5]
    crta5z = [z5,z6]
    crta6z = [z6,z7]
    
    
    fig = plt.figure()
    ax = fig.add_subplot(projection="3d")
    ax.plot(crta1x, crta1y, crta1z,color = "black")
    ax.plot(crta2x, crta2y, crta2z)
    ax.plot(crta3x, crta3y, crta3z)
    ax.plot(crta4x, crta4y, crta4z)
    ax.plot(crta5x, crta5y, crta5z)
    ax.plot(crta6x, crta6y, crta6z)
    plt.show()    
```
