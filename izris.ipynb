```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return x+2*x**2
```

```python
def izris(tabela):
    for i in tabela:
        def f(x):
            return -2*i**2+x*(4*i+1)
        plt.plot(np.linspace(-15,15,1000),f(np.linspace(-15,15,1000)),linewidth=0.5)
```

```python
plt.figure(figsize=(12,8))
izris(np.arange(-10,10,0.01))
#plt.plot(np.linspace(-15,15,1000),f1(np.linspace(-15,15,1000)),linewidth=2.5, color="red")
plt.savefig("slika")
```
