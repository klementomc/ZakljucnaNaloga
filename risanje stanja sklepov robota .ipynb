```python
from matplotlib import pyplot as plt
import numpy as np
from ipywidgets import interact


class Joint:
    def __init__(self, a, d, alpha):
        self.a = a
        self.d = d
        self.alpha = alpha


joint1 = Joint(0, 0.333, 0)
joint2 = Joint(0, 0, -np.pi/2)
joint3 = Joint(0, 0.316, np.pi/2)
joint4 = Joint(0.0825, 0, np.pi/2)
joint5 = Joint(-0.0825, 0.384, -np.pi/2)
joint6 = Joint(0, 0, np.pi/2)
joint7 = Joint(0.088, 0, np.pi/2)
Flange = Joint(0, 0.107, 0)

def DH(theta, alpha, d, a):
    st = np.sin(theta)
    ct = np.cos(theta)
    sa = np.sin(alpha)
    ca = np.cos(alpha)

    matrika = np.array([[ct, -st, 0, a],
                        [st*ca, ct*ca, -sa, -d*sa],
                        [st*sa, ct*sa, ca, d*ca],
                        [0, 0, 0, 1]])

    return matrika


#@interact(theta1=(-2.8973, 2.8973, 0.0001), theta2=(-1.7628, 1.7628, 0.0001), theta3=(-2.8973, 2.8973, 0.0001), theta4=(-0.0698, -3.0718, 0.0001), theta5=(-2.8973, 2.8973, 0.0001), theta6=(-0.0175, 3.7525, 0.0001), theta7=(-2.8973, 2.8973, 0.0001))
def plotting(theta1=(-2.8973, 2.8973, 0.0001), theta2=(-1.7628, 1.7628, 0.0001), theta3=(-2.8973, 2.8973, 0.0001), theta4=(-3.0718,-0.0698, 0.0001), theta5=(-2.8973, 2.8973, 0.0001), theta6=(-0.0175, 3.7525, 0.0001), theta7=(-2.8973, 2.8973, 0.0001)):

    plt.rcParams["figure.figsize"] = [15, 15]
    plt.rcParams["figure.autolayout"] = True

    T01 = DH(theta1, joint1.alpha, joint1.d, joint1.a)
    T12 = DH(theta2, joint2.alpha, joint2.d, joint2.a)
    T23 = DH(theta3, joint3.alpha, joint3.d, joint3.a)
    T34 = DH(theta4, joint4.alpha, joint4.d, joint4.a)
    T45 = DH(theta5, joint5.alpha, joint5.d, joint5.a)
    T56 = DH(theta6, joint6.alpha, joint6.d, joint6.a)
    T67 = DH(theta7, joint7.alpha, joint7.d, joint7.a)
    T78 = DH(0,0,0.107,0)

    t1 = T01
    t2 = np.matmul(T01, T12)
    t3 = np.matmul(t2, T23)
    t4 = np.matmul(t3, T34)
    t5 = np.matmul(t4, T45)
    t6 = np.matmul(t5, T56)
    t7 = np.matmul(t6, T67)
    t8 = np.matmul(t7, T78)

    x1, y1, z1 = t1[:-1, 3:][:]
    x2, y2, z2 = t2[:-1, 3:][:]
    x3, y3, z3 = t3[:-1, 3:][:]
    x4, y4, z4 = t4[:-1, 3:][:]
    x5, y5, z5 = t5[:-1, 3:][:]
    x6, y6, z6 = t6[:-1, 3:][:]
    x7, y7, z7 = t7[:-1, 3:][:]
    x8, y8, z8 = t8[:-1, 3:][:]

    x1 = x1[0]
    x2 = x2[0]
    x3 = x3[0]
    x4 = x4[0]
    x5 = x5[0]
    x6 = x6[0]
    x7 = x7[0]
    x8 = x8[0]

    y1 = y1[0]
    y2 = y2[0]
    y3 = y3[0]
    y4 = y4[0]
    y5 = y5[0]
    y6 = y6[0]
    y7 = y7[0]
    y8 = y8[0]

    z1 = z1[0]
    z2 = z2[0]
    z3 = z3[0]
    z4 = z4[0]
    z5 = z5[0]
    z6 = z6[0]
    z7 = z7[0]
    z8 = z8[0]

    crta1x = [0, x1]
    crta2x = [x1, x2]
    crta3x = [x2, x3]
    crta4x = [x3, x4]
    crta5x = [x4, x5]
    crta6x = [x5, x6]
    crta7x = [x6, x7]
    crta8x = [x7, x8]

    crta1y = [0, y1]
    crta2y = [y1, y2]
    crta3y = [y2, y3]
    crta4y = [y3, y4]
    crta5y = [y4, y5]
    crta6y = [y5, y6]
    crta7y = [y6, y7]
    crta8y = [y7, y8]

    crta1z = [0, z1]
    crta2z = [z1, z2]
    crta3z = [z2, z3]
    crta4z = [z3, z4]
    crta5z = [z4, z5]
    crta6z = [z5, z6]
    crta7z = [z6, z7]
    crta8z = [z7, z8]

    fig = plt.figure()
    ax = fig.add_subplot(projection="3d")
    ax.set_xlim(-1.5,1.5)
    ax.set_ylim(-1.5,1.5)
    ax.set_zlim(-1.5,1.5)
    ax.plot(crta1x, crta1y, crta1z, color="black")
    ax.plot(crta2x, crta2y, crta2z, color = "blue")
    ax.plot(crta3x, crta3y, crta3z, color = "red")
    ax.plot(crta4x, crta4y, crta4z, color = "green")
    ax.plot(crta5x, crta5y, crta5z, color = "yellow")
    ax.plot(crta6x, crta6y, crta6z, color = "purple")
    ax.plot(crta7x, crta7y, crta7z, color = "orange")
    ax.plot(crta8x, crta8y, crta8z)

    #print(t1)
    #print("----------------------")
    #print(t2)
    #print("----------------------")
    #print(t3)
    #print("----------------------")
    #print(t4)
    #print("----------------------")
    #print(t5)
    #print("----------------------")
    #print(t6)
    #print("----------------------")
    #print(t7)
    #print("----------------------")
    #print(t8)

    
    plt.show()

interact(plotting)
```
